```python
import librosa
import librosa.display
import IPython.display
from sklearn.preprocessing import scale
```

```python
wave, frequency = librosa.load("agua.wav", sr=None)
```

```python
wave
```

```python
frequency
```

```python
librosa.display.waveplot(wave, sr=frequency)
```

```python
IPython.display.Audio(wave, rate=frequency)
```

```python
mfccs = librosa.feature.mfcc(wave, sr=frequency)
mfccs.shape
```

```python
librosa.display.specshow(mfccs, sr=frequency, x_axis='time')
```
