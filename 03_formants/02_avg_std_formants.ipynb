```python
import csv
import numpy as np
```

```python
av_file = "av_f.csv"

theoretical_formants = {
    "a": (650, 1300),
    "e": (450, 1800),
    "i": (300, 2200),
    "o": (450, 1000),
    "u": (300, 1000),
}

observed_formants = {
    "a": [list(), list()],
    "e": [list(), list()],
    "i": [list(), list()],
    "o": [list(), list()],
    "u": [list(), list()],
}
observed_formants_std = {
    "a": [list(), list()],
    "e": [list(), list()],
    "i": [list(), list()],
    "o": [list(), list()],
    "u": [list(), list()],
}

with open(av_file) as av_f:
    reader = csv.DictReader(av_f)
    for line in reader:
        observed_formants[line["vowel"]][0].append(line["f1"])
        observed_formants[line["vowel"]][1].append(line["f2"])
        observed_formants_std[line["vowel"]][0].append(line["std_f1"])
        observed_formants_std[line["vowel"]][1].append(line["std_f2"])

print(observed_formants)
```

```python
for vowel in observed_formants.keys():
    f1_array = np.array(observed_formants[vowel][0]).astype(np.float)
    observed_formants[vowel][0] = np.nanmean(f1_array)
    f2_array = np.array(observed_formants[vowel][1]).astype(np.float)
    observed_formants[vowel][1] = np.nanmean(f2_array)

observed_formants
```

```python
for vowel in observed_formants_std.keys():
    std_f1_array = np.array(observed_formants_std[vowel][0]).astype(np.float)
    observed_formants_std[vowel][0] = np.nanmean(std_f1_array)
    std_f2_array = np.array(observed_formants_std[vowel][1]).astype(np.float)
    observed_formants_std[vowel][1] = np.nanmean(std_f2_array)

observed_formants_std
```
