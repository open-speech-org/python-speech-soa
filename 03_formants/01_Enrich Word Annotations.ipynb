```python
import os
import requests
import csv
```

```python
base_folder = "words_annotations"
annotations_folder = "annotations"
wav_folder = "wav"
annotations_file_path = os.path.join(base_folder, "word_map.txt")
with open(annotations_file_path) as af:
    annotations = af.readlines()
annotations
```

```python
ops_url = "http://localhost:8000"
audio_api_url = "/api/audio_data/{}/"
record_id_user_map = dict()
field_names = ["record_id", "transcription", "anonymous_name", "gender", "age", "accent", "pitch", "height", "weight"]
for line in annotations:
    record_id, annotation = line.replace("\n", "").split(",")
    response = requests.get(f"{ops_url}{audio_api_url.format(record_id)}")
    row = {
        "record_id": record_id,
        "transcription": annotation
    }
    if response.ok:
        profiles = response.json().get("anonymous_profile", [])
        if len(profiles) > 0:
            print(profiles[0].get("user", dict()))
            row = {**row, **profiles[0].get("user", dict())}
    record_id_user_map[record_id] = row
print(record_id_user_map)
```

```python
with open("av_f.csv") as tf, open('enriched_av_f.csv', 'w+') as eavf:
    reader = csv.DictReader(tf)
    field_names = ["record", "vowel", "f1", "f2", "d_f1", "d_f2", "std_f1", "std_f2", "intensity", "record_id", "transcription", "anonymous_name", "gender", "age", "accent", "pitch", "height", "weight" ]
    writer = csv.DictWriter(eavf, field_names, dialect='unix')
    writer.writeheader()
    for row in reader:
        new_row = {**row}
        if record_id_user_map.get(row['record'], False):
            new_row = {**new_row, **record_id_user_map[row['record']]}
        else:
            print(f"Row for {row['record']} does not exist")
        writer.writerow(new_row)
                  
```
