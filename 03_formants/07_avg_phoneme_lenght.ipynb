```python
import os
import numpy as np
import textgrids
```

```python
letters = "abcdefghijklmnopqrstuvwxyz"

lengths = dict()

base_folder = "words_annotations"
annotations_folder = "annotations"

for annotation in os.listdir(os.path.join(base_folder, annotations_folder)):

    text_grid = textgrids.TextGrid(os.path.join(base_folder, annotations_folder, annotation))
    intervals = text_grid[list(text_grid.keys())[0]]
    for interval in intervals:
        letter = interval.text.lower()
        if letter in letters:
            lengths.setdefault(letter, list()).append(interval.xmax - interval.xmin)

print(lengths)
```

```python
length_avg = dict()
length_std = dict()

for letter, lengths in lengths.items():
    nparray = np.array(lengths)
    length_std[letter] = nparray.std()
    length_avg[letter] = np.average(nparray)
```

```python
import pprint
pprint.pp(length_std)
pprint.pp(length_avg)
```
